- Alloted Set - It contains number of set from 1 to 5.

There are 5 different sections and 3 feature columns (given below) for every sections.

- Section **n** - Total questions attempted
- Wrong Answers **n** - Incorrect attempts

(n stands for section number)

```python
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.base import MIMEBase
from email import encoders

fromadd = "your own email id"
toadd = "other users email id"

msg = MIMEMultipart()

msg['From'] = fromadd
msg['To'] = toadd
msg['Subject'] = "MARKSHEET"

body = "With attachment (:"

msg.attach(MIMEText(body,'plain'))

filename = "table.txt"
attachment = open(filename,'rb')

part = MIMEBase('application','octate-stream')
part.set_payload((attachment).read())
encoders.encode_base64(part)

part.add_header('Content-Disposition',"attachment; filename = %s"%(filename))

msg.attach(part)

server = smtplib.SMTP('smtp.gmail.com', port=587)

server.starttls()

server.login(fromadd,"password")

text = msg.as_string()

server.sendmail(fromadd,toadd,text)

server.quit()
```
